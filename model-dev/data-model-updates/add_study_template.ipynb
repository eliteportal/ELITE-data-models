```python
import pandas as pd
import synapseclient as sc

syn = sc.login()
```

```python
new_template_name = "study_metadata_template"
```

```python
cols = [
    {"name": c["name"], "columnType": c["columnType"]}
    for c in syn.getColumns("syn54107382")
]
cols
cols = pd.DataFrame.from_records(cols)
cols = cols.rename(columns={'name': 'Attribute'})
cols = cols[cols["Attribute"] != "id"]
```

```python
term_df = pd.read_csv('../../data/new_terms/new_term_template.csv')
```

```python
new_term_df = pd.concat([term_df, cols])
new_term_df
```

```python
# set information
new_term_df = new_term_df.assign(
    Required=True,
    Parent="ManifestColumn",
    Properties="ManifestColumn",
    Source="sage_bionetworks",
    Ontology="sage_bionetworks",
    UsedIn=new_template_name,
)

new_term_df
```

```python
for i, r in new_term_df.iterrows(): 
    if r['columnType'] == 'BOOLEAN': 
        new_term_df.loc[i, 'Valid Values'] = 'true,false'

    if r['columnType'] == 'STRING' or r['columnType'] == 'STRING_LIST': 
        new_term_df.loc[i, 'Validation Rules'] = 'str'

    if r['Attribute'] in ['studyDescription', 'studyWiki', 'acknowledgement', 'methods', 'relatedStudies']: 
        new_term_df.loc[i, 'Description'] = "Synapse ID"

new_term_df
```

```python
new_attributtes = [
    {
        "Attribute": "dataCoordinationCenter",
        "Description": "Data coordination center the study is associated with",
        "Properties": "BaseAnnotation",
        "Parent": "ManifestColumn",
        "Source": "sage_bionetworks",
        "Ontology": "sage_bionetworks",
    }
]
```

```python
# add template attribute
new_template = [
    {
        "Attribute": new_template_name,
        "DependsOn": ",".join(
            new_term_df["Attribute"].to_list()
        ),  # not including component to prevent from being in DCA config to not confuse contributors
        "Description": "Template for adding study descriptions to a folder",
        "Properties": "Template",
        "Parent": "Component",
        "Source": "sage_bionetworks",
        "Ontology": "sage_bionetworks",
    }
]

new_template
```

```python
new_term_df = pd.concat([new_term_df, pd.DataFrame.from_records(new_template)])
new_term_df
```

```python
# write out new term to csv
new_term_df.to_csv('../../data/new_terms/study_metadata_template.csv', index = False)
```
