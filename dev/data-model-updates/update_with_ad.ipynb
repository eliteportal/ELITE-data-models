```python
import pandas as pd
```

```python
import numpy as np
```

```python
ad_dm = pd.read_csv(
    "https://raw.githubusercontent.com/adknowledgeportal/data-models/main/AD.model.csv"
)
ad_dm = ad_dm.set_index("Attribute")
ad_dm = ad_dm.drop_duplicates()
print(ad_dm.shape)
ad_dm.head()
```

```python
if ad_dm.index.is_unique:  # first check
    pass
else: 
    print('FAIL - Index is not unique')

ad_dm[ad_dm.index.duplicated(keep=False)].shape
temp =  ad_dm[ad_dm.index.duplicated(keep=False)].fillna('').astype(str).groupby("Attribute").agg(lambda x: ','.join(np.unique(x)).strip(','))
temp = temp.replace("", None)

if temp.index.is_unique: 
    print('PASS - Index is unique')
    pass
else:
    print("Index is not unique")

ad_dm = ad_dm[~ad_dm.index.duplicated(keep=False)]
ad_dm.shape
ad_dm = pd.concat([ad_dm, temp])

if ad_dm.index.is_unique: 
    print('PASS - Index is unique')
    pass
else:
    print("FAIL - Index is not unique")
```

```python
dm = pd.read_csv('../../EL.data.model.csv')
dm = dm.set_index('Attribute')
dm = dm.drop_duplicates()
print(dm.shape)
dm.head()
```

```python
dm.index.is_unique
```

```python
overlap = list(set(dm.index) & set(ad_dm.index))
dm.loc[overlap].head()
```

```python
dm.loc[overlap, "module"] = None
```

```python
dm.loc[overlap].shape
```

```python
dm = dm.fillna(ad_dm)
```
